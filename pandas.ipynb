### Create a subset

```python
subset_df = df[["title", "author", "sentence_str"]]
```

### Filter data

```python
df["school"].unique()
```

```python
criteria = "german_idealism"
```

```python
criteria == "german_idealism"
```

```python
criteria != "german_idealism"
```

```python
df["school"] == criteria
```

```python
idealism_filter = df["school"] == criteria
```

```python
idealism_df = df[idealism_filter]
```

```python
idealism_df["school"].unique()
```

```python
idealism_df["author"].unique()
```

```python
complex_filter = (df["school"] != "german_idealism") & (df["sentence_length"] > 300)
```

```python
final_df = df[complex_filter]
```

```python
print(df.shape)
print(final_df.shape)
```

## Visualize data
